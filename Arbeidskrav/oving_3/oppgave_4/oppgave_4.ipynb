$\rho=2 \cos (\phi)$ i kulekoordinater.

```python
# rho   rising
# phi   falling
# theta twizting
```

```python
import sympy as sp

# Define variables
rho, phi, theta = sp.symbols('rho phi theta')
x = rho * sp.sin(phi) * sp.cos(theta)
y = rho * sp.sin(phi) * sp.sin(theta)
z = rho * sp.cos(phi)

# Given equation
rho_eq = 2*sp.cos(phi)

# Substitute rho value in cartesian equations
x_cartesian = x.subs(rho, rho_eq)
y_cartesian = y.subs(rho, rho_eq)
z_cartesian = z.subs(rho, rho_eq)

# Plot
p = sp.plotting.plot3d_parametric_surface(x_cartesian, y_cartesian, z_cartesian, (phi, 0, sp.pi), (theta, 0, 2*sp.pi), 
                                          title=r"Plot of $\rho=2cos(\phi)$", xlabel="X", ylabel="Y")
#p.show()
```

$x^{2}+y^{2}=2 x-z^{2}$ i kartesiske koordinater.

```python
%reset -f

import sympy as sp

x, y, z = sp.symbols('x y z')

eq_2 = sp.Eq(x**2 + y**2,  sp.Rational(2)*x - z**2)
z = sp.solve(eq_2, z)

# Here we use plot_implicit to plot the equation in 3D space
p_2 = sp.plotting.plot3d(*z, (x, 0, 2), (y, -1, 1), title="Plot of $x^2 + y^2 = 2x - z^2$", 
                         xlabel="X", ylabel="Y", axis_center={'center', 'auto'}, 
                         xlim=(0,2), ylim=(-1,1), zlim=(-5,5))
```

$x^{2}+y^{2}=z^{2}-2 x$ i karteiske koordinater.

```python
%reset -f

import sympy as sp

x, y, z = sp.symbols('x y z')

eq_2 = sp.Eq(x**2 + y**2,  x**2-2*x)
x = sp.solve(eq_2, x)

# Here we use plot_implicit to plot the equation in 3D space
p_2 = sp.plotting.plot3d(*x, (z, -5, 5), (y, -5, 5), title="Plot of $x^2+y^2=z^2-2 x$", 
                         xlabel="X", ylabel="Y")
```

$r=2 \cos (\theta)$

```python
%reset -f

import sympy as sp

# Define variables
radius, theta, z = sp.symbols('rho phi theta')
x = sp.cos(theta) * radius
y = sp.sin(theta) * radius

# Given equation
radius_eq = 2*sp.cos(theta)

# Substitute rho value in cartesian equations
x_cartesian = x.subs(radius, radius_eq)
y_cartesian = y.subs(radius, radius_eq)
z_cartesian = z.subs(radius, radius_eq)

# Plot
p = sp.plotting.plot3d_parametric_surface(x_cartesian, y_cartesian, z_cartesian, (theta, 0, 2*sp.pi), 
                                          title=r"Plot of $r=2 \cos (\theta)$", xlabel="X", ylabel="Y")
#p.show()
```

$\rho=2 \sin (\phi)$ i kulekoordinater. Hint: Gjr om til kartesiske koordinater og se p hva du fr i xz-planet, dvs. nr y=0.

```python
%reset -f

import sympy as sp

#definer variables

rho, phi, theta = sp.symbols('rho phi theta')

x = rho * sp.sin(phi) * sp.cos(theta)
y = rho * sp.sin(phi) * sp.sin(theta)
z = rho * sp.cos(phi)

xyz = sp.Matrix((x, y, z))

rho_eq = 2 * sp.sin(theta)

p_6 = sp.plotting.plot3d_parametric_surface(*xyz.subs(rho, rho_eq), (phi, 0, sp.pi), (theta, 0, 2*sp.pi), 
                                            title=r"Plot of $\rho=2 \sin (\phi)$", xlabel="X", ylabel="Y")
```

$\sqrt{r^{2}+z^{2}}=2$ i sylinderkoordinater.

```python
%reset -f

import sympy as sp

radius, theta, z = sp.symbols('radius theta z')

# Given equation in cylindrical coordinates
eq_7 = sp.Eq(sp.sqrt(radius**2 + z**2), 2) 

radius_eq = sp.solve(eq_7, radius)

radius_eq
```

```python
positiv_lsning_radius_eq = radius_eq[1]

x = positiv_lsning_radius_eq * sp.cos(theta)
y = positiv_lsning_radius_eq * sp.sin(theta)

p_7 = sp.plotting.plot3d_parametric_surface(x, y, z, (theta, 0, 2*sp.pi), xlim=(-3, 3), ylim=(-3, 3), zlim=(-3, 3))
```
