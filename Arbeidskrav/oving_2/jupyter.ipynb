I denne oppgaven skal du regne ut tyngepunktet til en streng som ligger i planet. Skriv alle svar som tall, uten enheter.
Strengen har massetetthet $\delta=1$ og ligger langs kurven
$$
x(t)=3 \cdot \cos (t), y(t)=3 \cdot \sin (t), 0<t<\pi
$$
i planet.
Det første du kan gjøre er å finne ut hva slags kurve dette er og tegne en skisse.

```python
# Define the function
def oppg6(t):
    return 3 * np.cos(t), 3 * np.sin(t)

# Generate t values
t = np.linspace(0, np.pi, 1000)

# Generate x and y values
x, y = oppg6(t)

# Plot the function
plt.figure(figsize=(6,6))
plt.plot(x, y, label='oppg6(t)')
plt.title('Plot of oppg6(t)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.axis('equal')  # This ensures that the x and y scales are the same
plt.show()
```
